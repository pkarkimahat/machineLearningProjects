**Topic : Servo Mechaninism prediction**

**Import Library**

```python
import pandas as pd
import numpy as np
```

**Import dataset to read**

```python
SM= pd.read_csv('/content/Servo Mechanism.csv')
```

**Describe Data**

```python
SM.head()
```

```python
SM.tail()
```

```python
SM.describe()
```

```python
SM.shape
```

**Get Categories and Counts of Categorical Variables**

```python
SM[['Motor']].value_counts()
```

```python
SM[['Screw']].value_counts()
```

**Get encoding of categorical features**

```python
SM.replace({'Motor':{'A':0,'B':1,'C':2,'D':3,'E':4}},inplace=True)
SM.replace({'Screw':{'A':0,'B':1,'C':2,'D':3,'E':4}},inplace=True)
```

**Define Target Variable(y) and feature Variable(x)**


```python
y=SM['Class']
y.shape
y
```

```python
x=SM[['Motor', 'Screw', 'Pgain', 'Vgain']]
x=SM.drop('Class', axis=1)
x.shape
x
```

**Train test split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2529)
```

**Modeling**


```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

**Train Model**

```python
model.fit(x_train, y_train)
```

**Get Model Prediction**

```python
y_pred=model.predict(x_test)
y_pred.shape
y_pred

```

**Model Evaluation**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test,y_pred)
mean_absolute_error(y_test,y_pred)
r2_score(y_test,y_pred)
```

**Visualization of Actual and predited Results**

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs predicted")
plt.show()
```

**Get Future Prediction**

```python
x_new=SM.sample(1)
x_new
x_new.shape
x_new=x_new.drop('Class',axis=1)
x_new
x_new.shape
y_pred_new=model.predict(x_new)
y_pred_new
```
